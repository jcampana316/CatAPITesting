```python
import nltk
nltk.download('brown')
nltk.download('universal_tagset')

#importar el corpus y el modulod HMM
from nltk.corpus import brown
from nltk.tag import hmm

##variables segun clase ---> train_data -> sentences
# Tomar solo las primeras 5000 oraciones para el entrenamiento
sentences = brown.tagged_sents(tagset='universal')[:5000]

# Entrenar el modelo HMM
trainer = hmm.HiddenMarkovModelTrainer()
tagger = trainer.train_supervised(sentences)

#Oracion a etiquetar
sentence = ["El", "perro", "ladra"]

#Etiquetar usando el modelo entrenado (Viterbi)
tagged_sentence = tagger.tag(sentence)

#Imprimir resulado
print("Input: ", sentence)
print("Output con etiquetas: ", tagged_sentence)
```
